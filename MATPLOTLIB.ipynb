# Bar plot

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,70,60,82]
plt.xlabel('Coding Languages',fontsize=10)
plt.ylabel('Popularity in numbers',fontsize=10)
plt.title('Divya''s' 'Bar graph',fontsize=10)
cs=['y','r','g','m']
plt.bar(x,y,width=0.3,color='r',align='edge',edgecolor='black',linestyle=':',label='bar graph')
plt.bar(x,y,width=0.3,color='b',align='edge',edgecolor='black',linestyle=':',label='polularity')
#we can use align in center if we want.
plt.legend()
plt.show()
#(we can use alpha which is 0-1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Coding Languages',fontsize=10)
plt.ylabel('Popularity in numbers',fontsize=10)
plt.title('Divya''s' 'Bar graph',fontsize=10)

plt.bar(p,y,width,color='r',label='bar graph')
plt.bar(p1,z,width,color='b',label='polularity')

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel('Coding Languages',fontsize=10)
plt.ylabel('Popularity in numbers',fontsize=10)
plt.title('Divya''s' 'Bar graph',fontsize=10)

plt.barh(x,y,color='r',align='edge',edgecolor='black',linestyle=':',label='bar graph')
plt.barh(x,z,color='b',align='edge',edgecolor='black',linestyle=':',label='polularity')

plt.legend()
plt.show()
```

# SCATTER PLOT

```python
import matplotlib.pyplot as plt
x=['mon','tues','wed']
y=[1,2,3]

plt.scatter(x,y,color='r')
plt.title('Scatter plot',fontsize=10)
plt.xlabel('days',fontsize=10)
plt.ylabel('age of aunty',fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['mon','tues','wed']
y=[1,2,3]
#colors=['r','g','y']
colors=[10,20,30]
size=[100,200,300]

plt.scatter(x,y,s=size,alpha=0.191,marker='*',edgecolor='black',linewidth=4,cmap='BrBG')
t=plt.colorbar()
t.set_label('colorBar',fontsize=20)

plt.title('Scatter plot',fontsize=10)
plt.xlabel('days',fontsize=10)
plt.ylabel('age of aunty',fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=['mon','tues','wed']
y=[1,2,3]
z=[10,20,30]

colors=[10,20,30]
size=[100,200,300]

plt.scatter(x,y,s=size,alpha=0.191,marker='*',edgecolor='black',linewidth=4,cmap='BrBG')
plt.scatter(x,z,s=size,color='r',edgecolor='y',linewidth=4)

t=plt.colorbar()
t.set_label('colorBar',fontsize=20)

plt.title('Scatter plot',fontsize=10)
plt.xlabel('days',fontsize=10)
plt.ylabel('age of aunty',fontsize=10)

plt.show()
```

# MATPLOTLIB STACK PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,4,5,6]

area1=[1,2,3,4,4,5,6]
area=[1,2,4,6,8,12,2]
l=["area","area"]
plt.stackplot(x,area1,area,labels=l,colors=["m","black"],
             baseline="wiggle")# in baseline we can write zero, sym,wiggle
plt.legend(loc=2)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# STEP PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="python")#ms means marker size
plt.grid()
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.legend(loc=2)
plt.ylabel("y-axis")
plt.show()

```

# MATPLOTLIB FILL_BETWEEN PLOT

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

plt.plot(x, y, color="r")
plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.fill_between(x, y1=2, y2=4, color="g", where=(x >= 2) & (x <= 4), alpha=0.5)
plt.show()
```

# SAVEFIG 

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y,color='r')
plt.savefig("line",dpi=2000)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y,color='r')
plt.savefig("line45",dpi=2000,facecolor='g',
           transparent=False,bbox_inches="tight")
plt.grid()
plt.show()
```

# HOW TO ADD TEXT ON PLOT 

```python
import matplotlib.pyplot as plt

```

```python
x=[1,2,3,4,5]
y=[3,2,1,4,6]
plt.plot(x,y)
plt.title("Divya",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("python",fontsize=15)
plt.text(2,5,"java",fontsize=15,style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(3,1),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100),bbox={"facecolor":"red"})
plt.legend(["up"],loc=9,facecolor="red",edgecolor='y',framealpha=0.5,shadow=True)
plt.show()
```
